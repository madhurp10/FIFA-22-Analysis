### 1) Players Analysis on Age

```python
plt.figure(figsize=(22, 8))
ax = sns.countplot(x='age', data=df)
ax.set_title(label="Player Count Based on Age", fontsize=20)
ax.set_xlabel(xlabel="Age", fontsize=14)
ax.set_ylabel(ylabel="Number of Players", fontsize=14)
plt.show()
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(22,12))
sns.histplot(data=df, x='age', ax=ax[0, 0], kde=True).set(title="Age HistPlot")
sns.boxplot(data=df, x='age', ax=ax[0, 1]).set(title="Age Boxplot")
sns.scatterplot(data=df, x='age', y='overall', ax=ax[1, 0]).set(title="Rating wrt Age")
sns.scatterplot(data=df, x='age', y='pace', ax=ax[1, 1]).set(title="Pace wrt Age")
fig.tight_layout()
```

As we can see, there are not many top players below age 21. This is because in FIFA, at start young players are not given high rating but high potential.

21 - 30 is the peak of a Football player, and hence most of the players in this range are in the top 650 players.

## Top Players Analysis

```python
attributes = ['pace','dribbling', 'shooting', 'defending', 'passing' , 'physic']
```

```python
## Function which takes skills only DF, averages the data and converts it to a single row DF for pointplot
def calc_avg(df_skills):
    
    attributes = ['pace','dribbling', 'shooting', 'defending', 'passing' , 'physic']
    avg_skills_list = []
    
    for attribute in attributes:
        avg_skills = df_skills[attribute].mean()
        avg_skills_list.append(avg_skills) # List with the mean values of attributes

    avg_skills = pd.DataFrame(avg_skills_list).T ## Converting list to DF
    avg_skills.columns = attributes ## Adding Column names for the DF
    
    return avg_skills
```

### Top 10 Players

```python
## Top 10 (not including GK's)
top10 = mask.sort_values('overall', ascending = False).head(10)
top10
```

```python
top10_skills = pd.DataFrame(top10, columns=attributes)
top10_skills
```

```python
top10_avg = calc_avg(top10_skills)
top10_avg
```

```python
plt.figure(figsize=(20, 8))

sns.pointplot(data=top10_avg, color='green')
plt.title("Avg skills of Top 10 Players")
plt.xlabel("Skills", fontsize=20)
plt.ylabel("Value", fontsize=20)
plt.grid()
```
